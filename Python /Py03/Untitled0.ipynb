```python
# [[YOUR CODE]]
import math
```

```python
while True:
    while True:
        try:
            num1 = int(input("1번 정수를 입력하세요: "))
            break
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    while True:
        try:
            num2 = int(input("2번 정수를 입력하세요: "))
            break
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    sem = input("연산자를 입력하세요 (+, -, *, /, **): ")

    try:
        if sem == "+":
            print("결과:", num1 + num2)
        elif sem == "-":
            print("결과:", num1 - num2)
        elif sem == "*":
            print("결과:", num1 * num2)
        elif sem == "/":
            print("결과:", num1 / num2)
        elif sem == "**":
            print("결과:", math.pow(num1, num2))
        else:
            print("지원하지 않는 연산자입니다.")

    except ZeroDivisionError:
        print("잘못된 입력입니다. 0으로 나눌 수 없습니다.")

    print("-" * 20)
    retry = input("계속 계산하시겠습니까? (y/n): ")

    if retry == 'n':
        print("끝.")
        break
```
