```python
import random

class Account:
    account_count = 0
    
    def __init__(self, name, balance):
        self.bank = 'sc은행'
        self.name = name
        self.balance = balance

        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)
        
        self.account_num = str(num1) + " - " + str(num2) + " - " + str(num3)
        Account.account_count += 1

        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []

    def deposit(self, money):
        if money >= 1:
            self.balance = self.balance + money
            self.deposit_count = self.deposit_count + 1
            self.deposit_log.append(money)
            print(money, "원이 입금되었습니다.")
        else:
            print("입금은 1원 이상 가능합니다.")

    def withdraw(self, money):
        if money <= self.balance:
            self.balance = self.balance - money
            self.withdraw_log.append(money)
            print(money, "원이 출금되었습니다.")
        else:
            print("잔액이 부족합니다.")

    def display_info(self):
        print("은행이름:", self.bank)
        print("예금주:", self.name)
        print("계좌번호:", self.account_num)
        print("잔고:", format(self.balance, ","), "원")

# --- 여기서부터는 클래스 밖 (맨 왼쪽 벽에 붙여서!) ---
account_list = []
account_list.append(Account("파이썬", 1200000))
account_list.append(Account("어렵다", 500000))
account_list.append(Account("모두연", 2000000))
account_list.append(Account("아이펠", 2500000))

print("### 100만원 이상 고객 리스트 ###")
for acc in account_list:
    if acc.balance >= 1000000:
        acc.display_info()
```
