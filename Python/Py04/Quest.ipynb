Project.1 평행사변형, 사다리꼴 넓이를 구하는 메서드를 직접 작성할 수 있다.

```python
class square:
  def __init__(self):
     self.square = int(input("넓이를 구하고 싶은 사각형의 숫자를 써주세요. 1. 평행사변형 2. 사다리꼴"))

     if self.square == 1:
        print('평행사변형 함수는 par()입니다.')

     elif self.square == 2:
        print('사다리꼴 항수는 trape()입니다.')

     else:
        print('다시 입력해주세요')

  def par(self):
      width, height = map(int, input('가로, 높이를 입력하세요. 예시:가로,높이').split(','))
      area = width * height
      result = '평행사변형의 넓이는 : ' + str(area)
      return result

  def trape(self):
      width1, width2, height = map(int, input('가로1,가로2,높이를 입력하세요. 예시:가로1,가로2,높이').split(','))
      area = (width1 + width2) * height / 2
      result = '사다리꼴의 넓이는 : ' + str(area)
      return result

a = square()
```

```python
a.par()
```

```python
a.trape()
```

Project.2 Kiosk 클래스를 생성하고, 주문, 추가 주문 메서드를 적절하게 구현할 수 있다.
          지불, 주문표 작성 메서드를 적절하게 구현할 수 있다.    

```python
menu = ['Ame', 'latte', 'mocha', 'green_tea', 'smoothie', 'aid']
price = [2000, 3000, 4000, 3000, 4000, 4000]

class Kiosk:
  def __init__(self):
     self.menu = menu
     self.price = price

     self.order = []
     self.order_price = []

  def menu_print(self):
    for i in range(len(self.menu)):
      print(f'{i+1}. {self.menu[i]} : {self.price[i]}원')

  def menu_select(self):
     while True:
       n = int(input('음료번호를 고르세요 (종료는 0)): '))
       if n == 0:
          print("주문을 종료합니다.")
          break

       if 1 <= n <= len(self.menu):
           self.order.append(self.menu[n-1])
           self.order_price.append(self.price[n-1])
           print(f'{self.menu[n-1]} : {self.price[n-1]}원 담겼습니다.')
       else:
           print('없는 메뉴입니다. 다시 선택해주세요.')
           
  def pay(self):
     total = sum(self.order_price)
     print(f'지불할 금액은 {total}원 입니다.')

     method = int(input('결제수단은 선택하세요. (현금 1, 카드2) :'))
     if method == 1:
        print('직원을 호출하겠습니다.')

     elif method == 2:
        print('Ic칩 방향에 맞게 카드를 꽂아주세요.')
        print('*' * 20)

     else:
        print('다시 결제해주세요.')

  def table(self):
     for i in range(len(self.order)):
        print(f'{self.order[i]} : {self.order_price[i]}원')
     print('*' * 20)
     print(f'합계금액 : {sum(self.order_price)}원')
```

```python
a = Kiosk()
a.menu_print()  
a.menu_select()  
a.pay() 
a.table()
```
